#**Data** **Toolkit** (**Theory**)

```python
1. What is NumPy, and why is it widely used in Python?
NumPy is a Python library for working with numbers and arrays. It is fast, memory-efficient, and has many built-in functions for math, making it very useful in data science.


2. How does broadcasting work in NumPy?
Broadcasting means NumPy can perform operations on arrays of different shapes by automatically “stretching” them without making extra copies.

3. What is a Pandas DataFrame?
A DataFrame is like an Excel table in Python. It has rows and columns where you can store and analyze data.

4. Explain the use of the groupby() method in Pandas.
groupby() is used to split data into groups based on a column, then apply calculations like sum, mean, or count.

5. Why is Seaborn preferred for statistical visualizations?
Seaborn makes beautiful and easy-to-read charts, especially for statistics (like correlations, distributions, and trends).

6. What are the differences between NumPy arrays and Python lists?
NumPy arrays are faster and use less memory.

Lists are more flexible but slower.

NumPy arrays allow mathematical operations directly.

7. What is a heatmap, and when should it be used?
A heatmap is a colored chart that shows values in a grid. It’s good for visualizing correlations, patterns, or large data tables.

8. What does the term “vectorized operation” mean in NumPy?
Vectorized operation means performing calculations on entire arrays at once instead of looping through elements one by one.

9. How does Matplotlib differ from Plotly?

Matplotlib: Static charts (good for reports).

Plotly: Interactive charts (good for dashboards).

10. What is the significance of hierarchical indexing in Pandas?
It allows multiple levels of indexes in rows or columns, which makes working with multi-dimensional data easier.

11. What is the role of Seaborn’s pairplot() function?
pairplot() shows scatterplots between all column pairs and histograms for each column, useful for spotting relationships.

12. What is the purpose of the describe() function in Pandas?
describe() gives quick statistics like mean, median, min, max, and standard deviation for numeric data.

13. Why is handling missing data important in Pandas?
Missing data can give wrong results. Handling it (by filling, replacing, or removing) ensures correct analysis.

14. What are the benefits of using Plotly for data visualization?
Plotly makes interactive, zoomable, and dynamic charts that look modern and are good for presentations.

15. How does NumPy handle multidimensional arrays?
NumPy supports arrays with many dimensions (1D, 2D, 3D, etc.), which helps in working with images, matrices, or tensors.

16. What is the role of Bokeh in data visualization?
Bokeh is a Python library for interactive visualizations, often used for dashboards and web apps.

17. Explain the difference between apply() and map() in Pandas.

map() works on a single column/Series.

apply() works on entire DataFrames or multiple columns.

18. What are some advanced features of NumPy?

Linear algebra functions

Fourier transform

Random number generation

Broadcasting

19. How does Pandas simplify time series analysis?
Pandas has built-in tools for dates, resampling, rolling averages, and time-based indexing.

20. What is the role of a pivot table in Pandas?
A pivot table rearranges data, showing summaries like totals, averages, or counts grouped by categories.

21. Why is NumPy’s array slicing faster than Python’s list slicing?
Because NumPy arrays store data in continuous memory blocks, while lists store pointers, making them slower.

22. What are some common use cases for Seaborn?

Correlation heatmaps

Distribution plots

Pairwise plots

Regression plots

Categorical comparisons
```
