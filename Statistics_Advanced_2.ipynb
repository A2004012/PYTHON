1.What is hypothesis testing in statistics?

ans-Hypothesis testing in statistics is a method used to make inferences about a population based on sample data. It involves formulating a hypothesis about a population parameter and then using sample data to test whether there's enough evidence to support or reject that hypothesis.


2.What is the null hypothesis, and how does it differ from the alternative hypothesis?

ans- The null hypothesis (H0) is a statement of no effect or no difference, assuming that any observed difference in the data is due to chance. The alternative hypothesis (H1 or Ha) is the statement that there is an effect or a difference. The null hypothesis is tested against the alternative hypothesis using sample data.

3.Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.

ans- The significance level (α) in hypothesis testing is the probability of rejecting the null hypothesis when it is actually true (Type I error). Commonly set at 0.05, it determines the threshold for deciding whether the observed data provides enough evidence to reject the null hypothesis. If the p-value is less than α, the null hypothesis is rejected.

4.What are Type I and Type II errors? Give examples of each.

ans-

*   Type I error: Rejecting the null hypothesis when it is true. Example: Concluding a new medicine is effective when it's not.
*   Type II error: Failing to reject the null hypothesis when it is false. Example: Concluding a new medicine isn't effective when it actually is.




5.What is the difference between a Z-test and a T-test? Explain when to use each.

ans- Z-test: Used when the population variance is known and the sample size is large (typically n > 30), or when the data follows a normal distribution.

6: Write a Python program to generate a binomial distribution with n=10 and p=0.5, then plot its histogram.


```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
n = 10      # number of trials
p = 0.5     # probability of success
size = 1000 # number of samples

# Generate binomial distribution
data = np.random.binomial(n, p, size)

# Plot histogram
plt.hist(data, bins=range(n+2), edgecolor='black', alpha=0.7)
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.xlabel("Number of Successes")
plt.ylabel("Frequency")
plt.show()
```

 7: Implement hypothesis testing using Z-statistics for a sample dataset in Python. Show the Python code and interpret the results.
    - sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.2, 49.6, 50.1, 49.9, 50.8, 50.4, 49.9, 50.6, 50.0, 49.7, 50.2, 50.5, 50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9, 50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

```python
import numpy as np
from scipy import stats

# Sample data
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Hypothesized population mean
mu = 50

# Sample statistics
mean_sample = np.mean(sample_data)
std_sample = np.std(sample_data, ddof=1)
n = len(sample_data)

# Z-test
z_score = (mean_sample - mu) / (std_sample / np.sqrt(n))
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

print("Sample Mean:", mean_sample)
print("Z-Score:", z_score)
print("P-Value:", p_value)
```

8: Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Simulate normal distribution
data = np.random.normal(loc=50, scale=5, size=1000)

# Sample mean and std
mean = np.mean(data)
std_err = stats.sem(data)  # Standard error of the mean

# 95% Confidence Interval
confidence = 0.95
h = std_err * stats.t.ppf((1 + confidence) / 2., len(data)-1)
ci_lower, ci_upper = mean - h, mean + h

print("Sample Mean:", mean)
print("95% Confidence Interval:", (ci_lower, ci_upper))

# Plot histogram
plt.hist(data, bins=30, edgecolor='black', alpha=0.7)
plt.axvline(mean, color='red', linestyle='dashed', linewidth=2, label="Mean")
plt.axvline(ci_lower, color='green', linestyle='dashed', linewidth=2, label="95% CI Lower")
plt.axvline(ci_upper, color='green', linestyle='dashed', linewidth=2, label="95% CI Upper")
plt.title("Normal Distribution with 95% CI")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

9: Write a Python function to calculate the Z-scores from a dataset and visualize the standardized data using a histogram. Explain what the Z-scores represent in terms of standard deviations from the mean.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import zscore

# Generate random dataset
data = np.random.normal(50, 5, 1000)

# Calculate Z-scores
z_scores = zscore(data)

# Plot histogram of Z-scores
plt.hist(z_scores, bins=30, edgecolor='black', alpha=0.7)
plt.title("Z-Scores Histogram")
plt.xlabel("Z-Score")
plt.ylabel("Frequency")
plt.show()

print("Mean of Z-scores:", np.mean(z_scores))
print("Std Dev of Z-scores:", np.std(z_scores))
```
